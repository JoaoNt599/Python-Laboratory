```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 5, 8, 10, 12]
y = [2, 3, 7, 4, 6, 8]

titulo = "Gr치fico de Linhas"
eixox = "Eixo X"
eixoy = "Eixo Y"

# legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.plot(x, y, label = "Legenda", color = "green") # gr치fico de lihas
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 4] # Tamanho das barras

titulo = "Gr치fico de Barras"
eixox = "Eixo X"
eixoy = "Eixo Y"

# legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x, y, label = "Legenda") # gr치fico de barras
plt.legend()
plt.show()
plt.savefig("Figure1.png", dpi=400)
```

```python
x1 = [1, 5, 7, 9, 11]
y1 = [2, 3, 7, 1, 4]

x2 = [2, 6, 8, 10, 12]
y2 = [8, 1, 9, 2, 0]

titulo = "Comparando Gr치ficos"
eixox = "Eixo X"
eixoy = "Eixo Y"

# legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

# comparando gr치ficos
plt.bar(x1, y1, label = "Grupo 1")
plt.bar(x2, y2, label = "Grupo 2")
plt.legend()

plt.show()
# plt.savefig("Figure.pdf") # salvando figura
```

```python
x = [1, 5, 7, 9, 11]
y = [2, 3, 7, 1, 4]


titulo = "Gr치fico de Dispers칚o"
eixox = "Eixo X"
eixoy = "Eixo Y"

# legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x, y, label = "Pontos", marker="^", s=80, color = "green")
plt.plot(x, y, color = "black", linestyle="--")
plt.legend()

plt.show()
```
